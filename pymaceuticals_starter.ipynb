## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "resources/Mouse_metadata.csv"
study_results_path = "resources/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data = pd.merge(mouse_metadata, study_results, on="Mouse ID", how='outer')
# Display the data table for preview
combined_data
```

```python
# Checking the number of mice.
len(combined_data["Mouse ID"].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicates = combined_data[combined_data.duplicated(['Mouse ID', 'Timepoint'])]
duplicates["Mouse ID"]
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dup_mouse = combined_data[combined_data['Mouse ID'] == 'g989']
dup_mouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_data = combined_data
cleaned_data = cleaned_data[cleaned_data["Mouse ID"] != 'g989']
cleaned_data
```

```python
# Checking the number of mice in the clean DataFrame.
len(cleaned_data["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
grouped_drugs = cleaned_data.groupby(["Drug Regimen"])
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
drug_means = grouped_drugs["Tumor Volume (mm3)"].mean()
drug_median = grouped_drugs["Tumor Volume (mm3)"].median() 
drug_var = grouped_drugs["Tumor Volume (mm3)"].var() 
drug_std = grouped_drugs["Tumor Volume (mm3)"].std()
drug_sem = grouped_drugs["Tumor Volume (mm3)"].sem()

# Assemble the resulting series into a single summary dataframe.
summary_stats = pd.DataFrame({"Mean":drug_means,"Median":drug_median,"Variance":drug_var,"Standard Deviation":drug_std,"SEM":drug_sem})
summary_stats
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
statsdf = pd.DataFrame({"Mean":grouped_drugs["Tumor Volume (mm3)"].mean(),"Median":grouped_drugs["Tumor Volume (mm3)"].median(),"Variance":grouped_drugs["Tumor Volume (mm3)"].var(),"Standard Deviation":grouped_drugs["Tumor Volume (mm3)"].std(),"SEM":grouped_drugs["Tumor Volume (mm3)"].sem()})
statsdf
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of unique mice tested on each drug regimen using pandas.
drug_micecount = grouped_drugs['Mouse ID'].nunique()
mice_count_plot = drug_micecount.plot.bar(title="Mice per Drug Regimen", rot=45)
mice_count_plot.set_ylabel("Mouse Count")
```

```python
# Generate a bar plot showing the total number of unique mice tested on each drug regimen using pyplot.
drug_names = ["Capomulin",'Ceftamin','Infubinol', 'Ketapril', 'Naftisol', 'Placebo', 'Propriva', 'Ramicane', 'Stelasyn', 'Zoniferol']
plt.bar(drug_names, drug_micecount, align='center', width=.55)
plt.title("Mice per Drug Regimen")
plt.xlabel("Drug Regimen")
plt.ylabel("Mouse Count")
plt.xticks(rotation =45)
plt.show()
```
